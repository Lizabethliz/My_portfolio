### Matplotlib Charts

```python
import matplotlib.pyplot as plt
```

```python
a = [1, 2, 4, 7, 10, 12]
```

#### LINE CHART

```python
plt.plot(a)
plt.show()
```

```python
b = [10, 15, 20, 25, 30, 35]
```

```python
# Plottinrg (x, y) 
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b, "ro")
plt.show()         
```

```python
import numpy as np
```

```python
arr_one =  np.array(a)
arr_two =  np.array(b)
arr_three =  arr_two**2
arr_four = arr_one*2
```

```python
print(arr_one)
print(arr_two)
print(arr_three)
print(arr_four)
```

```python
plt.plot(arr_one, arr_two, "r--", arr_four, arr_three, "g^")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart')
plt.show()
```

### BAR CHART


```python
groups = ["A", "B", "C"]
x = [1, 2, 3]
```

```python
plt.bar(x, groups)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Chart')
plt.show()
```

```python
plt.bar(x, groups, color = "pink")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Chart')
plt.show()
```

```python
# Rotated/Horizontal Barchart
plt.barh(x, groups)
plt.xlabel('x axis')
plt.ylabel('Groups')
plt.title('Rotated bar Chart')
plt.show()
```

```python
plt.barh(x, groups, color = "pink")
plt.xlabel('x axis')
plt.ylabel('Groups')
plt.title('Rotated bar Chart')
plt.show()
```
