### QU 1 Suicide rate over the years based on sex? Draw a quantitative report from your analysis 

```python
sucide_rate_by_sex_year = df.groupby(['year', 'sex'], as_index=False)['suicides_no'].mean()
```

```python
sucide_rate_by_sex_year
```

```python
sns.lineplot(data = sucide_rate_by_sex_year, x = 'year', y = 'suicides_no', hue = 'sex')
plt.title('sucide_rate_by_sex_year')
plt.show()
```

### QU 2 Is age a considerable factor of committing suicide? Provide a report based on your analysis

```python
sucide_rate_by_age = df.groupby(['age'], as_index=False)['suicides_no'].mean()
```

```python
sucide_rate_by_age 
```

```python
sns.violinplot(data = df , x = 'suicides_no', y = 'age', palette='Set1')
plt.title('sucide_rate_by_age') 
plt.show()
```

### QU 3 The highest suicide rate happened in which of the generations?

```python
sucide_rate_by_generation = df.groupby(['generation'], as_index=False)['suicides_no'].mean()
```

```python
sucide_rate_by_generation
```

```python
sns.barplot(data = sucide_rate_by_generation , x='suicides_no', y='generation', color = '#FF8E00')
plt.title('sucide_rate_by_generation')
plt.show()
```

### QU 4 List the top 10 countries with the highest number of suicide rate?

```python
sucide_rate_by_countries = df.groupby(['country'], as_index=False)['suicides_no'].mean()
```

```python
sucide_rate_by_countries 
```

```python
sucide_rate_by_top_10_countries = sucide_rate_by_countries.sort_values(by = 'suicides_no', ascending=False).head(10)
```

```python
sucide_rate_by_top_10_countries 
```

```python
sns.barplot(data = sucide_rate_by_top_10_countries , x='suicides_no', y='country', color = '#FF8E00')
plt.title('sucide_rate_by_top_10_countries')
plt.show()
```

### QU 5 Does the degree of national wealth(gdp_per_capital) a considerable factor to suicide rate?

```python
suicide_rate_by_gdp_per_capita = df.groupby(['gdp_per_capita ($)'], as_index=False)['suicides_no'].mean()
```

```python
suicide_rate_by_gdp_per_capita
```

```python
suicide_rate_by_gdp_per_capita = suicide_rate_by_gdp_per_capita.sort_values(by = 'suicides_no', ascending=False).head(20)
```

```python
suicide_rate_by_gdp_per_capita
```

```python
sns.scatterplot(data = suicide_rate_by_gdp_per_capita, x ='gdp_per_capita ($)', y  ='suicides_no' , color ='#FF8E00')
plt.title('suicide_rate_by_gdp_per_capita')
plt.show()
```
