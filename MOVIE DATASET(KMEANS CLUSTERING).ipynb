### CLUSTERING WITH KMEANS

```python
from sklearn.cluster import KMeans
```

```python
ewscc = []
```

```python
for i in range(1, 150):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    ewscc.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 150), ewscc)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
accuracy_list = []
```

```python
for i in range(2, 150):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit_transform(x)
    a = silhouette_score(x, kmeans.labels_)
    print(i, a)
    accuracy_list .append(a)
```

```python
# Initializing Kmeans
kmeans = KMeans(n_clusters = 149, random_state = 20)
```

```python
kmeans.fit(x)
```

```python
plt.scatter(first_component, second_component, c = kmeans.labels_, alpha = 0.7)
plt.show()
```
