### DATA ENCODING OF CATEGORICAL COLUNMS
#### ONE HOT ENCODING

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
for x in categorical_columns:
    one_hot = pd.get_dummies(df[x], dtype = int)
    df = pd.concat([df,one_hot],axis = 1)
```

```python
df.head()
```

```python
df.drop(['Sex', 'Embarked', 'age_group','Age'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
# remove -inf fare
df['Fare'] = df['Fare'].astype('str')
df = df[df['Fare'] != '-inf']
df['Fare'] = df['Fare'].astype('float')
```

```python
x = df.drop('Survived', axis = 1)
y = df['Survived']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x['Fare'].unique()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 20)
```

### Kneighbor classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize the model
knn = KNeighborsClassifier(n_neighbors = 71)
```

```python
knn.fit(x_train, y_train)
```

### MODEL EVALUATION

```python
y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
accuracyList = []
n_neighbors = []
```

```python
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    a = accuracy_score(y_test, y_pred)
    print(i, a)
    accuracyList.append(a)
    n_neighbors.append(i)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(n_neighbors, accuracyList)
plt.ylabel('Accuracy score')
plt.xlabel('n_neighbors')
plt.title('KNN classifier on Titanic Dataset')
plt.show()
```
