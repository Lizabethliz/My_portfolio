```python
import pandas as pd
```

```python
outlet_small = pd.read_csv('outlet_size_small.csv')
outlet_medium = pd.read_csv('outlet_size_medium.csv')
oulet_high = pd.read_csv('outlet_size_high.csv')
```

### CONCAT() METHOD

```python
#Axis = 0 means join by row
#Axis = 1 means join by column
outlet_full = pd.concat([outlet_small, outlet_medium, oulet_high], axis = 0)
```

```python
data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

data2 = {
  "name": ["Sally", "Petef", "Micky"],
  "age": [77, 44, 22]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

## TYPES OF JOIN

### RIGHT JOIN

```python
df1.merge(df2, how = "right", on = "name")
```

### INNER JOIN

```python
df1.merge(df2, how = "inner", on = "name")
```

### OUTER JOIN

```python
df1.merge(df2, how = "outer", on = "name")
```

### OUTER JOIN

```python
df1.merge(df2, how = "right", on = "name")
```
