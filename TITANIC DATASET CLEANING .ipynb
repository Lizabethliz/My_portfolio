### FUTURE ENGINEERING A NEW COLUMN

```python
# Coverting age to age group
def age_converter(age):
    if age < 18:
        return 'Minor'
    elif age < 50:
        return 'Adult'
    else:
        return 'Old'
```

```python
# Checking it out
age_converter(40)
```

```python
# Checking it out
age_converter(60)
```

```python
df.Age.apply(age_converter)
```

```python
df['age_group'] = df.Age.apply(age_converter)
```
